### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # Different colors for each bar

# Plotting
plt.bar(categories, values, color=colors)

# Customization
plt.title('Bar Plot with Different Colors for Each Category')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show plot
plt.show()
```
